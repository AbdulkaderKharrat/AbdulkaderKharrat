```python
#numerical
#for each column
#n.1...> distrbution---->histogram and kde
#n.2...>outlayers---->boxplot
#n.3...>for all columns togather correlation between data--->heatmap
```

```python
# Assuming 'df_numerical_features' is a DataFrame with numerical columns.
# We will create a sample DataFrame to simulate this scenario.

import numpy as np

# Sample data creation
np.random.seed(0)

# Plotting Histogram with KDE and Boxplot for each numerical feature
fig, axes = plt.subplots(len(df_numerical_features.columns), 2, figsize=(12, 5 * len(df_numerical_features.columns)))

for i, col in enumerate(df_numerical_features.columns):
    # Histogram with KDE
    df_numerical_features[col].plot(kind='hist', density=True, ax=axes[i, 0], bins=20, alpha=0.5, color='blue')
    df_numerical_features[col].plot(kind='kde', ax=axes[i, 0], color='red')
    axes[i, 0].set_title(f'{col} - Histogram with KDE')

    # Boxplot
    df_numerical_features.boxplot(column=col, ax=axes[i, 1])
    axes[i, 1].set_title(f'{col} - Boxplot')

plt.tight_layout()
plt.show()
```

```python
from pickle import TRUE
correlation = df_numerical_features.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm' ,fmt = '.2f' , square= True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#for catagorical features
#unique values and values count & barplot & pieplot
```

```python
df_categorical_features['JobTitle'].unique()
print(len(df_categorical_features['JobTitle'].unique()))
```

```python
df_categorical_features['JobTitle'].value_counts()
```

```python
job_counts = df_categorical_features['JobTitle'].value_counts()

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart
axes[0].pie(job_counts, labels=job_counts.index, autopct='%1.1f%%', startangle=140)
axes[0].set_title('Job Title Distribution (Pie Chart)')

# Bar chart
job_counts.plot(kind='bar', ax=axes[1])
axes[1].set_title('Job Title Distribution (Bar Chart)')
axes[1].set_ylabel('Count')
axes[1].set_xlabel('Job Title')

plt.tight_layout()
plt.show()
```
